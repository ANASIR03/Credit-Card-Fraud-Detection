```python
import pandas as pd

training_data_34_66_path = '/data/notebook_files/training_data_34_66.csv'
training_data_10_90_path = '/data/notebook_files/training_data_10_90.csv'
X_test_path = '/data/notebook_files/X_test.csv'
y_test_path = '/data/notebook_files/y_test.csv'

training_data_34_66 = pd.read_csv(training_data_34_66_path)
training_data_10_90 = pd.read_csv(training_data_10_90_path)
X_test = pd.read_csv(X_test_path)
y_test = pd.read_csv(y_test_path)

describe_34_66 = training_data_34_66.describe()
describe_10_90 = training_data_10_90.describe()
describe_X_test = X_test.describe()
describe_y_test = y_test.describe()

describe_34_66, describe_10_90, describe_X_test, describe_y_test
```

```python
import matplotlib.pyplot as plt

def plot_class_distribution_bar(dataset_34_66, dataset_10_90, title):
    count_34_66 = dataset_34_66['Class'].value_counts(normalize=True)
    count_10_90 = dataset_10_90['Class'].value_counts(normalize=True)
    
    plt.bar(['34:66'], count_34_66[0], color='green')
    plt.bar(['34:66'], count_34_66[1], bottom=count_34_66[0], color='red')
    plt.bar(['10:90'], count_10_90[0], color='green')
    plt.bar(['10:90'], count_10_90[1], bottom=count_10_90[0], color='red')
    
    plt.ylabel('Proportion')
    plt.title(title)
    plt.legend(['Non-Fraud', 'Fraud'])
    plt.show()

plot_class_distribution_bar(training_data_34_66, training_data_10_90, 'Class Distribution Comparison')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

training_data_34_66 = pd.read_csv('/data/notebook_files/training_data_34_66.csv')
training_data_10_90 = pd.read_csv('/data/notebook_files/training_data_10_90.csv')

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(training_data_34_66['Amount'], bins=50, kde=True, color='blue', label='34:66')
sns.histplot(training_data_10_90['Amount'], bins=50, kde=True, color='orange', label='10:90')
plt.title('Distribution of Transaction Amount in Resampled Datasets')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(training_data_34_66['Time'], bins=50, kde=True, color='blue', label='34:66')
sns.histplot(training_data_10_90['Time'], bins=50, kde=True, color='orange', label='10:90')
plt.title('Distribution of Transaction Time in Resampled Datasets')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([training_data_34_66[training_data_34_66["Class"]==1]["Amount"], 
             training_data_34_66[training_data_34_66["Class"]==0]["Amount"]],
            labels=['Fraud (34:66)', 'Normal (34:66)'])
plt.title('Transaction Amounts in 34:66 Dataset')
plt.ylabel('Amount')
plt.yscale('log')

plt.subplot(1, 2, 2)
plt.boxplot([training_data_10_90[training_data_10_90["Class"]==1]["Amount"], 
             training_data_10_90[training_data_10_90["Class"]==0]["Amount"]],
            labels=['Fraud (10:90)', 'Normal (10:90)'])
plt.title('Transaction Amounts in 10:90 Dataset')
plt.ylabel('Amount')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
